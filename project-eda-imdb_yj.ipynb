```python
# trim whitespace from content ratings

movies['content_rating'] = movies['content_rating'].astype(str).str.strip()
```

#### Convert the following content ratings to "UNRATED": NOT RATED, APPROVED, PASSED, GP.

```python
# Answer:

import numpy as np

unrated_list = ['NOT RATED', 'APPROVED', 'PASSED', 'GP']

movies['content_rating'] = np.where(movies['content_rating'].isin(unrated_list),
                                     'UNRATED',
                                     movies['content_rating'])
```

```python
movies['content_rating'].value_counts(dropna=False)
```

#### Convert the following content ratings to "NC-17": X, TV-MA.

```python
# Answer:

nc17_list = ['X', 'TV-MA']

movies['content_rating'] = np.where(movies['content_rating'].isin(nc17_list),
                                     'NC-17',
                                     movies['content_rating'])
```

```python
movies['content_rating'].value_counts(dropna=False)
```

#### Calculate the average star rating for movies 2 hours or longer, and compare that with the average star rating for movies shorter than 2 hours.

```python
# Answer:
# The average star rating for movies 2 hours or longer is slightly higher 
# than that for movies shorter than 2 hours (7.95 compared to 7.84)

movies_2hlonger = movies.loc[movies['duration']>=120]
print(movies_2hlonger['star_rating'].mean())

movies_2hshorter = movies.loc[movies['duration']<120]
print(movies_2hshorter['star_rating'].mean())
```

#### Use a visualization to detect whether there is a relationship between duration and star rating.

#### Figure out something "interesting" using the actors data!

```python
# Clean up actor names in the actors_list column

movies['clean_actors'] = movies['actors_list'].apply(lambda x: x[0:-1].split(','))
```

```python
movies['clean_actors'] = movies['clean_actors'].apply(lambda cell: [x[2:] for x in cell])
```

```python
movies['clean_actors'] = movies['clean_actors'].apply(lambda cell: [x.strip("'") for x in cell])
```

```python
movies['clean_actors']
```

```python
# Join all actor names into a single list

result = sum(movies['clean_actors'].tolist(), [])
```

```python
# Count occurrences of each name in the list and sort in descending order
# Robert De Niro has acted in the most movies in the dataset

from collections import Counter

result2 = Counter(result)
result2.most_common()
```
